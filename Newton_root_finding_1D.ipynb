```python
# multimodal function
from numpy import sin
from numpy import arange
import numpy as np
import matplotlib.pyplot as plt
 
```

```python
# objective function
def objective(x):
 return -x * sin(x)
```

```python
def plotFunction(): 
    # define range for input
    r_min, r_max = 0.0, 10.0
    # sample input range uniformly at 0.1 increments
    inputs = arange(r_min, r_max, 0.1)
    # compute targets
    results = objective(inputs)
    # create a line plot of input vs result
    plt.plot(inputs, results)

    # Suggestion for font type in plots from: 
    # https://matplotlib.org/stable/gallery/text_labels_and_annotations/text_fontdict.html
    font = {'family': 'serif',
            'color':  'black',
            'weight': 'normal',
            'size': 16,
            }

    plt.title('Function $f\,(x) = -t \,\sin(x)$ ', fontdict=font)
    plt.xlabel('$x$', fontdict=font)
    plt.ylabel('$f\,(x)$', fontdict=font)

    # define optimal input value
#    x_optima = 7.9787
    # draw a vertical line at the optimal input
 #   plt.axvline(x=x_optima, ls='--', color='gray')
    # show the plot
    return plt
```

```python
# Show function 
plt = plotFunction()
plt.show()
```

```python
# Derivative of the objective function 
def dfdx(x):    
    return -np.sin(x) - x*np.cos(x)
```

```python
# Newton's root-finding algorithm
x = 5                 # Initial value 
f = objective(x)      # Evaluate function at the initial value 
g = -6                # Goal value of f(x) 
step_scale = 0.1     # Lambda (step scale factor)  
while abs(f - g) > 0.1: 
    s = dfdx(x)                             # Slope of the objective function at local x 
    x = x + step_scale * (g - f) * (1/s)    # Updated location following the slope
    f = objective(x)                        # Updated function value 
    print("x = ", x, "f(x) = ", f)    
```

```python
plt = plotFunction()
# optimal input value when approaching from the left 
x_optima_left = x
# draw a vertical line at the optimal input
plt.axvline(x=x_optima_left, ls='--', color='red')
# show the plot

# draw a horizontal line at the optimal point
plt.axhline(y=g, ls='--', color='gray')

plt.show()
```
